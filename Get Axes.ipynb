```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(np.linspace(0.0, 1.0, 50), np.linspace(0.0, 1.0, 50))
Z = 1 - (X + Y) / 2
ax.plot_surface(X, Y, Z, cmap=matplotlib.cm.Blues)
ax.set_zbound(lower=0.0, upper=1.0)
plt.savefig("plane")
```
